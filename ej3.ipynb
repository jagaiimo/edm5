```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import PartialDependenceDisplay

# Load the dataset
df = pd.read_csv("kc_house_data.csv")

# Take a random sample to avoid performance issues with large data
sample_size = 5000  # Adjust as needed
df_sample = df.sample(n=sample_size, random_state=42)

# Define the features and target variable
features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'yr_built']
target = 'price'

X_sample = df_sample[features]
y_sample = df_sample[target]

# Train a Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_sample, y_sample)

# Prepare subplots grid: 2 rows x 3 columns for 6 features
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Flatten the axes array for compatibility with PartialDependenceDisplay
axs = axs.flatten()

# Plot Partial Dependence
PartialDependenceDisplay.from_estimator(
    model,
    X_sample,
    features=features,
    ax=axs,
    kind="average"
)

plt.tight_layout()
plt.show()





```
